#### Carregando os dados
O comando pd.read_csv é um comando da biblioteca pandas (pd.) e carrega os dados do arquivo csv indicado para um objeto *dataframe* do pandas.

```python
renda = pd.read_csv('./input/previsao_de_renda.csv')
renda.drop(axis=1,columns=['Unnamed: 0'],inplace=True)
```

#### Entendimento dos dados - Univariada
Nesta etapa tipicamente avaliamos a distribuição de todas as variáveis. 

```python
prof = ProfileReport(renda, explorative=True, minimal=True)
prof
```

```python
def quantitativas_continuas(x):
    print('Variável: ' + x)
    print(renda[x].describe())
    print('\n')

    fig, axs = plt.subplots(ncols = 2, figsize = (15,6))
    sns.distplot(renda[x], ax = axs[0])
    sns.boxplot(data = renda, y = x, palette = 'Blues_r', ax = axs[1])
    
def quantitativas_discretas(x):
    print('Variável: ' + x)
    print(renda[x].describe())
    print('\n')

    fig, axs = plt.subplots(ncols = 2, figsize = (15,6))
    sns.countplot(renda[x], ax = axs[0])
    sns.boxplot(data = renda, y = x, palette = 'Blues_r', ax = axs[1])
    
def qualitativas(x, y):
    print('Variável: ' + x)
    print(renda[x].value_counts(normalize = True).round(2))
    print('\n')
    print('Variável: ' + y)
    print(renda[y].value_counts(normalize = True).round(2))
    print('\n')

    fig, axs = plt.subplots(ncols = 2, figsize=(15,6))
    sns.countplot(renda[x], ax = axs[0], order = renda[x].value_counts().index, palette = 'Blues_r')
    sns.countplot(renda[y], ax = axs[1], order = renda[y].value_counts().index, palette = 'Greens_r')
    
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    data_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return data_out

def correlacao(w, x, y):
    print('Tabela de correlação:')
    print(renda[[w, x, y, 'total']].corr())
    print('\n')
    
    fig, axs = plt.subplots(ncols = 3, figsize=(24,6))
    sns.scatterplot(data = renda, x = w, y = 'total', ax = axs[0], color = 'Blue')
    sns.scatterplot(data = renda, x = x, y = 'total', ax = axs[1], color = 'Green')
    sns.scatterplot(data = renda, x = y, y = 'total', ax = axs[2], color = 'Red')
```
